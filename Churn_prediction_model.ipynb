```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from imblearn.combine import SMOTEENN
import pickle
```

```python
df = pd.read_csv('df_churn.csv')
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0', axis = 1)	
```

**Creating X and Y variables**

```python
x = df.drop('Churn', axis = 1)
```

```python
y = df['Churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

**Decision Tree Classifier**

```python
model_dt = DecisionTreeClassifier(criterion = 'gini', random_state = 100, max_depth = 6, min_samples_leaf = 8) 
```

```python
model_dt.fit(x_train, y_train)
```

```python
y_pred = model_dt.predict(x_test)
```

```python
y_pred
```

```python
print(classification_report(y_test, y_pred, labels = [0,1]))
```

```python
print(confusion_matrix(y_test, y_pred, labels = [0, 1]))
```

```python
sm = SMOTEENN()
x_resampled, y_resampled = sm.fit_resample(x, y)
```

```python
xr_train, xr_test, yr_train, yr_test = train_test_split(x_resampled, y_resampled, test_size = 0.2)
```

```python
model_dt_smote = DecisionTreeClassifier(criterion = 'gini', random_state = 100, max_depth = 6, min_samples_leaf = 8) 
```

```python
model_dt_smote.fit(xr_train, yr_train)
```

```python
yr_pred_smote = model_dt_smote.predict(xr_test)
```

```python
print(classification_report(yr_test, yr_pred_smote, labels = [0,1]))
```

```python
print(confusion_matrix(yr_test, yr_pred_smote, labels = [0, 1]))
```

**Randon Forest Classifier**

```python
model_rf = RandomForestClassifier(criterion = 'gini', random_state = 100, max_depth = 6, min_samples_leaf = 8) 
```

```python
model_rf.fit(x_train, y_train)
```

```python
y_pred_rf = model_rf.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_rf, labels = [0,1]))
```

```python
x_resampled, y_resampled = sm.fit_resample(x, y)
```

```python
xr_train, xr_test, yr_train, yr_test = train_test_split(x_resampled, y_resampled, test_size = 0.2)
```

```python
model_smote_rf = RandomForestClassifier(criterion = 'gini', random_state = 100, max_depth = 6, min_samples_leaf = 8) 
```

```python
model_smote_rf.fit(xr_train, yr_train)
```

```python
yr_pred_smote = model_smote_rf.predict(xr_test)
```

```python
print(classification_report(yr_test, yr_pred_smote, labels = [0,1]))
```

```python
print(confusion_matrix(yr_test, yr_pred_smote, labels = [0, 1]))
```

**Saving the model**

```python
filename = 'RFC_model.sav'
```

```python
pickle.dump(model_smote_rf, open(filename, 'wb'))
```

**Load model**

```python
load_model = pickle.load(open(filename, 'rb'))
```

```python
load_model.score(xr_test, yr_test)
```
